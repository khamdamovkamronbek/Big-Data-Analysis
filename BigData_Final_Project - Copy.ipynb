```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the datasets
github_data = pd.read_csv(r"C:\Users\hamda\Downloads\students_mental_health_survey.csv")
survey_data = pd.read_csv(r"C:\Users\hamda\Downloads\students_mental_health_survey.csv")
github_data.head()
```

```python
# Survey dataset preview
survey_data.head()
```

```python

github_data.columns = github_data.columns.str.strip()
print(github_data.columns)
```

```python
# Missing values
print(github_data.isnull().sum())
print(survey_data.isnull().sum())
```

```python
# GPA vs Stress Level visualization
sns.scatterplot(data=survey_data, x='Stress_Level', y='CGPA')
plt.title('Stress Level vs GPA')
plt.show()
```

```python
# Mental Health Issues in GitHub dataset
sns.countplot(data=github_data, x='Stress_Level')
plt.title('Mental Health Issue Count')
plt.show()
```

```python
#Students who sleep and exercise more show better academic performance.
github_data['CGPA'] = pd.to_numeric(github_data['CGPA'], errors='coerce')
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Sleep_Quality', y='CGPA', data=github_data)
plt.title('CGPA vs Sleep Duration')
plt.xlabel('Sleep Duration Category')
plt.ylabel('CGPA')
plt.show()
```

## Conclusion:
- Higher stress is associated with lower GPA.
- Students who sleep/exercise more perform better.
- GitHub data supports Inha survey findings.
